```python
metric = "laplacian_variance"

# Imposta la dimensione della figura
plt.figure(figsize=(12, 8))  # Aumenta la larghezza e l'altezza del grafico

# Crea il grafico
plt.subplot(2, 2, 1)  # Posiziona il grafico nella prima posizione della griglia 2x2

# Distribuzione per il dataframe degradato
sns.histplot(df_degradato[metric], kde=True, label='Degradato', color='red', bins=30)

# Distribuzione per il dataframe non degradato
sns.histplot(df_non_degradato[metric], kde=True, label='Non Degradato', color='blue', bins=30)

# Aggiungi il titolo e la legenda
plt.title(f'Distribuzione della {metric.replace("_", " ").title()}')
plt.legend()

# Mostra il grafico
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 12))  # Imposta un grafico più grande

# Box plot orizzontale per 'laplacian_variance' (primo grafico)
plt.subplot(2, 1, 1)  # Posizione superiore della griglia 2x1
sns.boxplot(data=df_non_degradato[['laplacian_variance']], orient='h', color='blue', label='Non Degradato')
plt.title('Distribuzione Orizzontale della Laplacian Variance Del Val Non-Degradato')
plt.legend()

# Box plot orizzontale per 'gradient_mean' (secondo grafico)
plt.subplot(2, 1, 2)  # Posizione inferiore della griglia 2x1
sns.boxplot(data=df_degradato[['laplacian_variance']], orient='h', color='red', label='Degradato')
plt.title('Distribuzione Orizzontale della Laplacian Variance Del Degradato')
plt.legend()

# Mostra i grafici
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = "/mnt/data/val_000027.jpg"
image = cv2.imread(image_path)

# Convert to grayscale for noise analysis
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Compute the histogram to understand the noise distribution
hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

# Display the image and histogram
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(gray_image, cmap='gray')
axes[0].set_title("Grayscale Image")
axes[0].axis("off")

axes[1].plot(hist)
axes[1].set_title("Histogram of Pixel Intensities")
axes[1].set_xlabel("Pixel Value")
axes[1].set_ylabel("Frequency")

plt.show()
```
